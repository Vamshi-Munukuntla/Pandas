```python
df.drop(columns='Remarks', inplace=True)
```

```python
df.sample()
```

```python
df.set_index('Sr No', inplace=True)
```

```python
df.sample()
```

```python
dicts = {'Date dd/mm/yyyy':'Date',
        'Startup Name':'Startup',
        'Industry Vertical': 'Vertical',
        'SubVertical':'Sub_Vertical',
        'City  Location':'Location',
        'Investors Name':'Investors',
        'InvestmentnType':'Investment_Round',
        'Amount in USD': 'Amount'}
```

```python
df.rename(columns=dicts, inplace=True)
```

```python
df.Amount.fillna(0, inplace=True)
```

```python
df.info()
```

```python
df.Amount = df.Amount.str.replace(',','')
df.Amount = df.Amount.str.replace('undisclosed','0')
df.Amount = df.Amount.str.replace('unknown','0')
df.Amount = df.Amount.str.replace('Undisclosed','0')
df.Amount = df.Amount.str.replace('14342000+','14342000')
df.Amount = df.Amount.str.replace(r'[^0-9.]', '', regex=True)
```

```python
df['Amount'] = df['Amount'].astype(str).fillna('0')
```

```python
df['Amount'] = df['Amount'].astype(float)
```

```python
df.info()
```

```python
def to_inr(dollar):
    inr = dollar*82.5
    return round(inr/10000000,2)
```

```python
df['Amount'] = df['Amount'].apply(to_inr)
```

```python
df.sample()
```

```python
df['Date'] = df['Date'].str.replace('05/072018', '05-07-2018')
df['Date'] = df['Date'].str.replace('13/04.2015', '13-04-2015')
df['Date'] = df['Date'].str.replace('01/07/015', '01-07-2015')
df['Date'] = df['Date'].str.replace(r'\\xc2\\xa0', '', regex=True)
df['Date'] = df['Date'].str.replace('12/05.2015', '12-05-2015')
df['Date'] = df['Date'].str.replace('15/01.2015', '15-01-2015')
df['Date'] = df['Date'].str.replace('22/01//2015', '22-01-2015')
df['Date'] = df['Date'].str.replace(r'\\\\xc2\\\\xa010/7/2015', '10-7-2015', regex=True)
```

```python
df['Date'] = pd.to_datetime(df.Date,format='%d-%m-%Y')
```

```python
df.Amount = df.Amount.fillna(0)
```

```python
df = df.dropna(subset=['Date', 'Startup', 'Vertical', 'Location', 'Investors', 'Investment_Round', 'Amount'])
```
